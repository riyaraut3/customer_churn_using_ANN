```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df = pd.read_csv("/content/Churn_Modelling.csv")
```

```python
#Divide the features in dependent and independent features
x = df.iloc[:, 3:13]
y = df.iloc[:, 13]
```

```python
x
```

```python
# Feature engineering
geography_dummies = pd.get_dummies(x["Geography"])
geography_dummies = geography_dummies.astype(int)
```

```python
geography_dummies
```

```python
gender = pd.get_dummies(x["Gender"])
gender = gender.astype(int)
```

```python
gender
```

```python
x =pd.concat([x, gender, geography_dummies], axis =1)
```

```python
x =x.drop(columns ="Geography")
x = x.drop(columns ="Gender")
```

```python
# Splitting the dataset into training set and test set
from sklearn.model_selection import train_test_split

X_train,X_test,y_train, y_test = train_test_split(x,y,test_size=0.2, random_state=1)
```

```python
X_train
```

**🔁 fit_transform(X_train) vs transform(X_test) — Why the difference?**

> 🧪 fit_transform(X_train):

> fit: Learns the scaling parameters (like mean and standard deviation for StandardScaler) from the training data only.

> transform: Applies those learned parameters to scale X_train.

> 🧪 transform(X_test):
Only uses the parameters learned from X_train (mean, std, min, max, etc.) to scale X_test.

No fitting is done on test data, to avoid data leakage.

```python
X_train
```

```python
# standard sclar - standardization or normalization of data
from sklearn.preprocessing import StandardScaler

sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

```python
import tensorflow
import keras
```

```python
from tensorflow.keras.models import Sequential
# Sequential: A linear stack of layers. You can build a model layer by layer
# by simply adding layers using .add().

from tensorflow.keras.layers import Dense
# Dense: A fully connected (dense) layer where every neuron is connected
# to every neuron in the previous layer. Used for most neural network layers.

from tensorflow.keras.layers import LeakyReLU, PReLU, ELU, ReLU
# Activation layers:
#   - ReLU (Rectified Linear Unit): Outputs max(0, x). Fast and widely used.
#   - LeakyReLU: Similar to ReLU but allows a small gradient for x < 0
#     (helps avoid "dying ReLU" problem).
#   - PReLU (Parametric ReLU): Like LeakyReLU but with a learnable negative slope.
#   - ELU (Exponential Linear Unit): Similar to ReLU but smooths the transition
#     for negative values (can improve training in some cases).

from tensorflow.keras.layers import Dropout
# Dropout: Regularization layer that randomly drops a fraction of neurons
# during training to prevent overfitting.
```

```python
# Initialize the ANN
classifier = Sequential()  # Create a linear stack of layers

# Input layer and first hidden layer
classifier.add(Dense(units=13, activation="relu", input_dim=13))
# 'units=13': number of neurons (same as number of input features)
# 'input_dim=13': explicitly specify input shape (required for first layer)

# Second hidden layer
classifier.add(Dense(units=7, activation="relu"))

# Third hidden layer
classifier.add(Dense(units=6, activation="relu"))

# Output layer units 1 as we want either 1 or 0
classifier.add(Dense(units=1, activation="sigmoid"))
# 'sigmoid': used for binary classification (output between 0 and 1)
```

```python
classifier.compile(optimizer="adam", loss="binary_crossentropy", metrics=["accuracy"])
```

We used binary cross-entropy as the loss function because our task is binary classification—predicting between two classes (e.g., 0 or 1). It measures how well the predicted probabilities match the actual class labels. The Adam optimizer was chosen for its efficiency and adaptive learning rate, which speeds up training and often performs better out of the box. Our input layer has 13 neurons to match the 13 input features, and the output layer has 1 neuron with a sigmoid activation to output a probability for the positive class. We use accuracy as a metric to monitor how often the model correctly predicts the class labels.

```python
optimi = tensorflow.keras.optimizers.Adam(learning_rate=0.01)
```

make classifier using .add() --> .compile()--> .fit(x_train,y_train) --> we can do early stopping if we want to update in such a way that if the model weights are not updating at all then

```python
callback = keras.callbacks.EarlyStopping(monitor='loss')
```

```python
model_history = classifier.fit(X_train,y_train, validation_split=0.33, batch_size =10, epochs=100, callbacks =callback )
```

```python
model_history.history.keys()
```

```python
# make prediction and evaliate the model
y_pred = classifier.predict(X_test)
y_pred = (y_pred>0.5).astype("int32")
```

```python
y_pred
```

```python
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test,y_pred)
```

```python
cm
```

```python
from sklearn.metrics import accuracy_score
score = accuracy_score(y_test, y_pred)
score
```

```python
from sklearn.metrics import f1_score, precision_score, recall_score
f1_score = f1_score(y_test, y_pred)
precision_score = precision_score(y_test, y_pred)
recall_score = recall_score(y_test, y_pred)

```
